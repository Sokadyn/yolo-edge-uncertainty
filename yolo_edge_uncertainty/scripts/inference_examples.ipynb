```python
import os
import sys

# allow importing from parent directory
current_folder = globals()['_dh'][0]
data_location = os.path.join(current_folder,'data.csv')
sys.path.append(os.path.abspath(os.path.join(os.path.dirname(data_location), '../..')))

import pandas as pd
from matplotlib import pyplot as plt

pd.set_option('display.max_columns', None)
from utlis_results import *

from ultralytics import settings
settings.update({"sync": False, "runs_dir": "../inference_examples"})
from ultralytics import YOLOEdgeUncertainty, YOLO

import cv2
```

```python
model = YOLO('yolo11n.pt')
dataset_examples = '../dataset_examples/'
img = 'oxford_town_centre_4227.jpg'
img_path = os.path.join(dataset_examples, img)
results = model.predict(img_path, plots=True, save=True, verbose=False, exist_ok=True, name='yolo11n')
save_dir = results[0].save_dir

img_bgr = cv2.imread(os.path.join(save_dir, img))
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```
